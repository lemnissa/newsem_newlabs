```python
import numpy as np
import matplotlib.pyplot as plt
import math

def plotting(file_name, ax, color = 'r', dot = 'x'):
    data = np.genfromtxt(file_name, delimiter=',', names=['n', 't'])
    print(data['t'])    
    
    x = []
    y = []
    for i in range(len(data['t'])):
        if data['t'][i] != 0:
            y.append(data['t'][i])
            x.append(data['n'][i])
    '''
    if file_name == "bubble_sort_o-0.txt":
        x = x[2:-3]
        y = y[2:-3]
    if file_name == "bubble_sort_o-2.txt":
        x = x[:4]
        y = y[:4]
    if file_name == "bubble_sort_o-1.txt":
        x = x[:4]
        y = y[:4]
    if file_name == "bubble_sort_o-3.txt":
        x = x[3:3] + x[5:-2] + x[-1:]
        y = y[3:3] + y[5:-2] + y[-1:]
    '''
    if file_name[0] == 'q' or file_name[0] == 'm' or file_name[0] == 'i'or file_name[0] == 'b':
        x = x[:-1]
        y = y[:-1]
    x = [math.log(each) for each in x]
    y = [math.log(each) for each in y]

    z = np.polyfit(x, y, 1)
    p = np.poly1d(z)
    x_p = np.linspace(min(x), max(x))
    
    ax.plot(x_p, p(x_p), c=color, ls='-', label=str(file_name[:-4]) + '() ' + 'N = ' + str(round(p[1], 1)))# + ', C = '  + str(math.exp(p[0]))[:4]+str(math.exp(p[0]))[-4:] )
    ax.plot(x, y, dot, c=color)

```
